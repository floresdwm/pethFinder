```python
import pandas as pd
import requests
import os
import re
from urllib.parse import urlparse
from io import BytesIO
from PIL import Image

# Função para baixar a imagem e salvá-la localmente
def download_image(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)
        return True
    else:
        return False

# Função para substituir caracteres inválidos nos nomes de arquivo do Windows
def sanitize_filename(filename):
    # Substituir caracteres inválidos por "_"
    return re.sub(r'[<>:"/\\|?*]', '_', filename)

# Ler o arquivo Excel
excel_file = "databases_clean.xlsx"
df = pd.read_excel(excel_file)

# Criar a pasta para salvar as imagens se não existir
output_folder = "imagens"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterar sobre as linhas do DataFrame
for index, row in df.iterrows():
    # Extrair informações
    abrigo = row['Abrigo']
    endereco_abrigo = row['Endereço do Abrigo']
    url = row['URL da Imagem']
    url_id = row['URL']

    # Extrair o ID da URL
    parsed_url = urlparse(url_id)
    url_id = os.path.splitext(os.path.basename(parsed_url.path))[0]

    # Formatar o nome do arquivo
    filename = f"{abrigo}_{endereco_abrigo}_id_{url_id}.jpg"
    filename = os.path.join(output_folder, sanitize_filename(filename))

    # Baixar a imagem e salvar localmente
    download_image(url, filename)

    print(f"Imagem baixada e salva como {filename}")

print("Download de imagens concluído!")
```
